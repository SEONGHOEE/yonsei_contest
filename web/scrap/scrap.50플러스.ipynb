```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# 50plus 교육 목록 URL (AI 검색 결과)
BASE_URL = "https://50plus.or.kr/education.do?cost=ALL&state=ALL&type=ALL&term=AI&page="

opts = Options()
opts.add_argument("--window-size=1280,1200")
driver = webdriver.Chrome(options=opts)

detail_urls = []

try:
    for page in range(1, 34):   # 1~33페이지 반복
        url = BASE_URL + str(page)
        print("페이지 이동:", url)
        driver.get(url)
        time.sleep(1.5)

        # 신청마감/신청하기 버튼 추출
        buttons = driver.find_elements(By.CSS_SELECTOR, "td.td-normal-btn a")
        print("링크 개수:", len(buttons))

        # 현재 페이지 10개 URL을 모두 수집
        for btn in buttons:
            href = btn.get_attribute("href")
            if href:
                detail_urls.append(href)

    print("총 수집된 URL 개수:", len(detail_urls))

finally:
    driver.quit()

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import csv

opts = Options()
opts.add_argument("--window-size=1280,1200")
driver = webdriver.Chrome(options=opts)

results = []  # title, description, url 저장 리스트

try:
    for i, url in enumerate(detail_urls, 1):
        driver.get(url)
        time.sleep(1.2)

        # 1) 제목(title) 추출
        title = driver.find_element(By.CSS_SELECTOR, "h2.show-title").text

        # 2) 내용(description) 추출
        try:
            # "내용 :" 라벨 찾기
            label = driver.find_element(By.XPATH, "//label[contains(text(), '내용')]")
            parent_td = label.find_element(By.XPATH, "./..")  # 부모 요소(td)

            full_text = parent_td.text.strip()
            description = full_text.replace("내용 :", "").strip()

        except:
            description = ""   # 내용 없는 페이지 예외 처리

        # 3) 저장
        results.append([title, description, url])

        print(f"{i}/{len(detail_urls)} 완료")
        print("제목 :", title)
        print("description :", description[:50], "...")  # 앞 50자 미리보기
        print()

finally:
    driver.quit()


# CSV 저장
with open("서울50플러스_최종.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["title", "description", "url"])
    writer.writerows(results)

print("수집 완료!")
```
